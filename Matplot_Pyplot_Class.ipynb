# Matplotlib - Understanding the Figure Object

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
# Data 
a = np.linspace(0,10,11)
```

```python
b = a ** 4
```

```python
a
```

```python
b
```

```python
x = np.arange(0,10)
```

```python
y = 2*x
```

```python
x
```

```python
y
```

```python
fig = plt.figure()

# Large Axis

axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)

axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel('X1 Label')
axes1.set_ylabel('Y1 Label')
axes1.set_title('Fig 1')

# Small Axis
axes2 = fig.add_axes([0.1,0.4,0.5,0.5])
axes2.plot(a,b)

axes2.set_xlim(1,2)
axes2.set_ylim(0,50)
axes2.set_xlabel('X2 Label')
axes2.set_ylabel('Y2 Label')
axes2.set_title('Fig 2')
plt.show()
```

# Matplotlib - Figure Parameters

```python
fig = plt.figure(dpi=100)
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)

fig.savefig('dpifig.png',bbox_inches='tight')
plt.show()
```

# Matplotlib - Subplots Functionality

```python
fig,axes = plt.subplots(nrows=3,ncols=1)
for ax in axes:
    ax.plot(x,y)
fig.suptitle('Main Title')
#plt.tight_layout() this is to adjust the spacing between the graphs automatically

fig.subplots_adjust(wspace=1,hspace=0.5)

```

# Matplotlib Styling - Legends

```python
x = np.linspace(0,11,10)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x,label='X vs X')

ax.plot(x,x**2,label='X vs X**2')
# ax.legend(loc=2) # this legend function adds the details of lines on left top of graph
ax.legend(loc=(0.1,0.5))
plt.show()
```

# Matplotlib Styling - Colors and Styles

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x= np.linspace(0,10,11)
```

```python
x
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x,color='#e6c017',label='X VS X')

ax.plot(x,x+1,color='red',label='X VS X+1')
ax.legend()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x,color='#e6c017',linewidth=2,ls='--',marker='o',markersize='10',markerfacecolor='red',
       markeredgewidth=4,markeredgecolor='orange') # ls is line stayle like dash

ax.plot(x,x+1,color='red',label='X VS X+1')
```
