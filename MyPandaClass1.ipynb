# Joins in Pandas

```python
# three typw of joing 

# 1 - Inner
# 2 - Outer
# 3 - Left or Right Join
import pandas as pd 
import numpy as np
```

```python
registrations = pd.DataFrame({'reg_id':[1,2,3,4],'name':['Andrew','Bobo','Claire','David']})
logins = pd.DataFrame({'log_id':[1,2,3,4],'name':['Xavier','Andrew','Yolanda','Bobo']})
```

```python
registrations
```

```python
logins
```

```python
pd.merge(registrations,logins,how='inner',on='name')
```

```python
pd.merge(registrations,logins,how='left',on='name')
```

```python
pd.merge(registrations,logins,how='right',on='name')
```

```python
pd.merge(registrations,logins,how='outer',on='name')
```

```python
pd.merge(registrations,logins,how='outer',left_on='name',right_on='name')
```

# Resample

A common operation with time series data is resampling based on the time series index. Let's see how to use the resample() method. [reference]

```python
sales = sales.set_index('DATE')
```

```python
sales.resample(rule='A').mean() # resample is the same as GROUP BY in rule we define 'A' means we try to run GROUP BY on
# year bcz its a DATETIME object and as we know with GROUP BY agregation function is important so we used mean 
```

When calling .resample() you first need to pass in a rule parameter, then you need to call some sort of aggregation function.

The rule parameter describes the frequency with which to apply the aggregation function (daily, monthly, yearly, etc.)
It is passed in using an "offset alias" - refer to the table below. [reference]

The aggregation function is needed because, due to resampling, we need some sort of mathematical rule to join the rows (mean, sum, count, etc.)
